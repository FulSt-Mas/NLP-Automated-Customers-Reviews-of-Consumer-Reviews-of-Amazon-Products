```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the path to your dataset
data_path = "/Users/massih/Code/Ironhack/1_Project/5_Project NLP | Automated Customers Reviews/NLP Automated Customers Reviews of Consumer Reviews of Amazon Products/artifacts/transformer_artifacts_1_data_preprocessing/preprocessed_reviews_transformer.csv"

# Load the dataset from the provided path
df = pd.read_csv(data_path)

# Ensure that `cleaned_text` column is of string type (just in case)
df['cleaned_text'] = df['cleaned_text'].astype(str)

# Calculate the length of each text entry in `cleaned_text`
df['text_length'] = df['cleaned_text'].apply(len)

# Compute statistics on the text lengths
average_length = df['text_length'].mean()
max_length = df['text_length'].max()
min_length = df['text_length'].min()
mode_length = df['text_length'].mode()[0]
length_distribution = df['text_length'].describe()

# Display the histogram of text lengths
plt.figure(figsize=(12, 6))
sns.histplot(df['text_length'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Text Lengths')
plt.xlabel('Text Length')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Display a box plot of text lengths
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['text_length'], color='lightgreen')
plt.title('Box Plot of Text Lengths')
plt.xlabel('Text Length')
plt.grid(True)
plt.show()

# Display the results
print("Text Length Statistics:")
print(f"Average text length: {average_length}")
print(f"Maximum text length: {max_length}")
print(f"Minimum text length: {min_length}")
print(f"Most common text length (mode): {mode_length}")
print("\nDistribution Summary:")
print(length_distribution)

# Optional: Display the DataFrame with the new `text_length` column
df.head()  # Display the first few rows to verify the new column
```
