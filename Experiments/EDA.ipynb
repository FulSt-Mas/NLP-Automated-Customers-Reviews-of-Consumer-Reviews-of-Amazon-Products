```python
import pandas as pd

# Load the data
def load_data(file_path):
    return pd.read_csv(file_path)

# Get main info about the data
def get_data_info(data):
    print("\n-------- Quick Overview --------")
    print(data.head())  # Print the first few rows
    print("\n-------- Summary --------")
    print(data.info())  # Print a concise summary of the DataFrame
    print("\n-------- Description --------")
    print(data.describe())  # Print statistical summary

# Usage
file_path = '../Data/Reviews_of_Amazon_Products_sm.csv'
data = load_data(file_path)
get_data_info(data)
```
